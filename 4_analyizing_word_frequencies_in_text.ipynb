```python
# Load text file

with open("alice_in_wonderland.txt", "r") as file:
    text = file.read()
```

```python
# Preprocess the text

text = text.lower()
text = re.sub(r'[^\w\s]', '' , text)
tokens = word_tokenize(text)
```

```python
# Display sample tokens

print(tokens[:20])
```

```python
# Calculate word frequency

word_counts = Counter(tokens)

print("Most common words:", word_counts.most_common(10))
```

```python
# Calculate basic statistcal measures

import numpy as np

# Convert word frequenciew to a list of counts
word_frequencies = list(word_counts.values())

mean_freq = np.mean(word_frequencies)  # OR sum(word_frequencies) / len(word_frequencies)
variance_freq = np.var(word_frequencies) # OR sum((word_length - mean_freq) ** 2  for word_length in word_frequencies)
std_dev_freq = np.std(word_frequencies)

print(f"Mean Frequency: {mean_freq}")
print(f"Variance: {variance_freq}")
print(f"Standard Deviation: {std_dev_freq}")
```
